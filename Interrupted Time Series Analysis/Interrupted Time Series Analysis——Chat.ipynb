## 导入所需的 Python 库 + 合并文件

```python
import pandas as pd  # 用于数据处理
import matplotlib.pyplot as plt  # 用于绘图
import statsmodels.api as sm  # 用于统计建模
import seaborn as sns  # 更美观的绘图

# =============== 1. 加载数据 ================
# 请注意：你需要先把两个原始文件合并为一个名为 combined_nonstimulant_ADHD_data.csv 的文件，或者在这里加载两个 CSV 并手动合并。
# 示例格式为：每行一个月的数据，列包括 date, actual_cost, items, quantity, drug
# 加载两个文件
df1 = pd.read_csv("Atomoxetine hydrochloride (0404000S0)-spending-by-all_practices.csv")
df2 = pd.read_csv("Guanfacine (0404000V0)-spending-by-all_practices.csv")

# 加药品名称列
df1["drug"] = "Atomoxetine"
df2["drug"] = "Guanfacine"

# 合并
combined = pd.concat([df1, df2])
combined.to_csv("combined_nonstimulant_ADHD_data.csv", index=False)
combined = pd.read_csv("combined_nonstimulant_ADHD_data.csv")
combined["date"] = pd.to_datetime(combined["date"])  # 将日期列转换为时间格式

# =============== 2. 计算单位处方成本 ================
combined["cost_per_item"] = combined["actual_cost"] / combined["items"]

# =============== 3. 绘图：每月处方数量变化趋势图 ================
plt.figure(figsize=(12, 6))
sns.lineplot(data=combined, x="date", y="items", hue="drug", marker="o")
plt.axvline(pd.to_datetime("2023-05-01"), color="red", linestyle="--", label="Atomoxetine Shortage Start")
plt.axvline(pd.to_datetime("2023-09-01"), color="orange", linestyle="--", label="Guanfacine Shortage Start")
plt.title("每月非兴奋型 ADHD 药物的处方数量变化趋势图")
plt.ylabel("处方总数量（items）")
plt.legend()
plt.tight_layout()
plt.show()

# =============== 4. 绘图：单位处方成本变化趋势图 ================
plt.figure(figsize=(12, 6))
sns.lineplot(data=combined, x="date", y="cost_per_item", hue="drug", marker="o")
plt.title("每个处方单位的平均成本变化趋势图")
plt.ylabel("每个处方成本 (£)")
plt.xlabel("日期")
plt.legend()
plt.tight_layout()
plt.show()

# =============== 5. 中断时间序列分析：整体分析图 ================
# 画出一张图，展示干预前后的预测线段 + 回归拟合结果

def plot_itsa_fitted(df, drug_name, intervention_date):
    df = df[df["drug"] == drug_name].copy()
    df = df[df["date"] >= "2022-10-01"]
    df["time"] = range(1, len(df)+1)
    df["intervention"] = (df["date"] >= intervention_date).astype(int)
    df["time_after"] = (df["time"] - df[df["intervention"] == 1]["time"].min() + 1) * df["intervention"]
    
    X = sm.add_constant(df[["time", "intervention", "time_after"]])
    model = sm.OLS(df["items"], X).fit()
    df["predicted"] = model.predict(X)

    plt.figure(figsize=(10, 5))
    plt.plot(df["date"], df["items"], label="实际值", marker="o")
    plt.plot(df["date"], df["predicted"], label="ITS 拟合值", linestyle="--")
    plt.axvline(pd.to_datetime(intervention_date), color="red", linestyle="--", label="短缺开始")
    plt.title(f"{drug_name} 处方数量的 ITS 模型拟合图")
    plt.xlabel("日期")
    plt.ylabel("处方总数量")
    plt.legend()
    plt.tight_layout()
    plt.show()
    return model

# 画出类似论文参考图的拟合线
model_atom_fitted = plot_itsa_fitted(combined, "Atomoxetine", "2023-05-01")
model_guanf_fitted = plot_itsa_fitted(combined, "Guanfacine", "2023-09-01")

# =============== 6. 中断时间序列分析（Atomoxetine） ================
df_atom = combined[combined["drug"] == "Atomoxetine"].copy()
df_atom = df_atom[df_atom["date"] >= "2022-10-01"]
df_atom["time"] = range(1, len(df_atom)+1)
df_atom["intervention"] = (df_atom["date"] >= "2023-05-01").astype(int)
df_atom["time_after"] = (df_atom["time"] - df_atom[df_atom["intervention"] == 1]["time"].min() + 1) * df_atom["intervention"]

X_atom = sm.add_constant(df_atom[["time", "intervention", "time_after"]])
y_atom = df_atom["items"]
model_atom = sm.OLS(y_atom, X_atom).fit()

print("\nAtomoxetine 中断时间序列分析结果:")
print(model_atom.summary())

# =============== 7. 中断时间序列分析（Guanfacine） ================
df_guanf = combined[combined["drug"] == "Guanfacine"].copy()
df_guanf = df_guanf[df_guanf["date"] >= "2022-10-01"]
df_guanf["time"] = range(1, len(df_guanf)+1)
df_guanf["intervention"] = (df_guanf["date"] >= "2023-09-01").astype(int)
df_guanf["time_after"] = (df_guanf["time"] - df_guanf[df_guanf["intervention"] == 1]["time"].min() + 1) * df_guanf["intervention"]

X_guanf = sm.add_constant(df_guanf[["time", "intervention", "time_after"]])
y_guanf = df_guanf["items"]
model_guanf = sm.OLS(y_guanf, X_guanf).fit()

print("\nGuanfacine 中断时间序列分析结果:")
print(model_guanf.summary())

# =============== 8. 成本对比汇总表 ================
summary = combined.groupby("drug").agg({
    "actual_cost": "sum",       # 总花费
    "items": "sum",             # 总处方数量
    "quantity": "sum",          # 总药品数量
    "cost_per_item": "mean"     # 平均每个处方的成本
}).reset_index()

print("\n两个非兴奋型药物的成本对比汇总:")
print(summary)  # 打印对比表格
```
